```python
import pandas as pd
a= pd.read_csv('f10.csv')
b= pd.read_csv('f14.csv')
c= pd.concat([a, b])
c
```

```python
d=c.drop_duplicates('tweet',keep='last').dropna(subset=['tweet'])
d
```

```python
e=list()
f=d.values.tolist()
for x in f:
    if x[2].count('#') <= 4:
        e.append(x)
from pandas import DataFrame
g=DataFrame(e,columns=['0','user','tweet','time','keyword'])
g=g.drop(['0'], axis=1)
g.to_csv('g.csv')
g
```

```python
g['time']=pd.to_datetime(g['time'])
```

```python
g = g.sort_values(by='time')
```

```python
g.reset_index(drop=True)
```

```python
g.to_csv('final_dataset.csv')
```
